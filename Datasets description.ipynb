Для данной лабораторной работы я выбрал следующие датасеты:

##1. Регрессия
---
###Источник: https://www.kaggle.com/datasets/mashlyn/online-retail-ii-uci
---
###Название: Online Retail II (данные о розничных онлайн-продажах)
---
###Описание

Для задачи регрессии был выбран датасет Online Retail II, содержащий реальные транзакционные данные интернет-магазина за определённый период. Каждая запись в датасете представляет собой отдельную покупку и включает следующие параметры: идентификатор товара, описание, количество единиц, цену за штуку, страну покупателя и дату совершения заказа.

На основе этих данных удобно формулировать задачу регрессии, направленную на прогнозирование стоимости покупки и исследование поведения покупателей.

Практическая значимость задачи заключается в возможности использования модели для оценки потенциальной выручки, выявления закономерностей в покупательской активности, оптимизации ассортимента товаров и повышения эффективности маркетинговых стратегий.

Датасет обладает чёткой структурой и содержит достаточное количество числовых и категориальных признаков, что делает его удобным для применения стандартных методов регрессии.

##2. Классификация.
---
###Источник: https://www.kaggle.com/datasets/dineshmanikanta/machine-failure-predictions/data
---
###Название: Machine Failure Predictions (предсказание поломки станка)
---
###Описание

Для решения задачи классификации был выбран датасет, содержащий сенсорные данные, полученные со станков в процессе их работы. В каждой записи представлены параметры текущего состояния оборудования: температура воздуха, температура технологического процесса, скорость вращения, крутящий момент и степень износа инструмента. Также указан факт возникновения отказа в данный момент времени и тип произошедшей неисправности.

Подобные данные широко применяются службами эксплуатации промышленных предприятий. Их ключевая задача заключается в своевременном обнаружении возможных отказов, снижении простоя оборудования и уменьшении затрат на ремонт.

Методы машинного обучения позволяют эффективно анализировать такие данные, выявлять скрытые закономерности и прогнозировать вероятность возникновения различных типов неисправностей.

Кроме того, я работаю в сфере АСУ ТП, поэтому использование данного датасета способствует развитию практических навыков, непосредственно связанных с анализом технологических процессов и мониторингом состояния оборудования.
